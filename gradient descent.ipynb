```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def function(x):
    return(x+3)**2
```

```python
def gradient(x):
    return 2*(x+3)
```

```python
def gradient_descent(starting_point,learning_rate,num_iterations):
    x = starting_point
    histroy = []
    for _ in range(num_iterations):
        grad = gradient(x)
        x = x-learning_rate*grad
        histroy.append(x)
        if abs(grad)<1e-6:
            break
    return x,histroy

starting_point = 2
learning_rate = 0.1
num_iterations = 50
```

```python
final_x,histroy = gradient_descent(starting_point,learning_rate,num_iterations)
print(f"Local minimum occurs at: x={final_x}")
```

```python
x_values = np.linspace(-6,3,100)
y_values = function(x_values)
plt.plot(x_values,y_values,label='y=(x+3)**2')
plt.scatter(histroy,function(np.array(histroy)),color='red',label='path taken by Gradient Descent')
plt.title('Gradient Descent:')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
